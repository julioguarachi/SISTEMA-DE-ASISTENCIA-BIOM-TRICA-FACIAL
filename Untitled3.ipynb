```python
# cargado de imagenes
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import time
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Configuración para optimizar el rendimiento en CPU
tf.config.threading.set_intra_op_parallelism_threads(4)
tf.config.threading.set_inter_op_parallelism_threads(2)

# =====================================================
# PARÁMETROS DEL MODELO
# =====================================================
IMG_SIZE = 64
BATCH_SIZE = 8
LATENT_DIM = 100
EPOCHS = 200
CHANNELS = 3

# =====================================================
# RUTAS DE LAS CARPETAS
# =====================================================
#malla_path = "/content/drive/MyDrive/ROSTROS/malla" --eliminar
img_sf_path = "/content/drive/MyDrive/ROSTROS/img_sf"


# =====================================================
# FUNCIÓN PARA CARGAR IMÁGENES
# =====================================================
#def load_images_from_folder(folder_path, target_size=(IMG_SIZE, IMG_SIZE), max_images=200):
def load_images_from_folder(folder_path, target_size=(IMG_SIZE, IMG_SIZE), max_images=1000):
    """
    Carga imágenes desde una carpeta
    Args:
        folder_path: Ruta de la carpeta
        target_size: Tamaño de redimensionamiento
        max_images: Número máximo de imágenes a cargar
    Returns:
        numpy array con las imágenes
    """
    images = []
    valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp')

    # Verificar si la carpeta existe
    if not os.path.exists(folder_path):
        print(f"⚠️ La carpeta {folder_path} no existe")
        return np.array([])

    # Obtener lista de archivos de imagen
    files = [f for f in os.listdir(folder_path) if f.lower().endswith(valid_extensions)]
    print(f"📁 Encontrados {len(files)} archivos en {os.path.basename(folder_path)}")

    # Limitar el número de imágenes
    files_to_load = files[:max_images]

    for filename in files_to_load:
        img_path = os.path.join(folder_path, filename)
        try:
            # Cargar y preprocesar imagen
            img = Image.open(img_path)

            # Convertir a RGB si es necesario
            if img.mode != 'RGB':
                img = img.convert('RGB')

            # Redimensionar
            img = img.resize(target_size, Image.Resampling.LANCZOS)

            # Normalizar a [-1, 1]
            img_array = np.array(img, dtype=np.float32) / 127.5 - 1.0
            images.append(img_array)

        except Exception as e:
            print(f"❌ Error al cargar {filename}: {e}")
            continue

    return np.array(images)

# =====================================================
# CARGA DE IMÁGENES DESDE AMBAS CARPETAS
# =====================================================
print("=" * 60)
print("🚀 INICIANDO CARGA DE IMÁGENES")
print("=" * 60)

all_images = []
# Cargar imágenes de rostros sin fondo
if os.path.exists(img_sf_path):
    img_sf_images = load_images_from_folder(img_sf_path)
    if len(img_sf_images) > 0:
        all_images.extend(img_sf_images)
        print(f"✅ Rostros sin fondo cargados: {len(img_sf_images)}")
    else:
        print(f"⚠️ No se encontraron imágenes en {img_sf_path}")
else:
    print(f"⚠️ La carpeta {img_sf_path} no existe")

# =====================================================
# VERIFICACIÓN Y CREACIÓN DE DATASET SINTÉTICO
# =====================================================
if len(all_images) == 0:
    print("\n⚠️ No se encontraron imágenes reales. ")
else:
    print(f"\n✅ Total de imágenes reales cargadas: {len(all_images)}")

# Convertir a numpy array
dataset = np.array(all_images, dtype=np.float32)
print(f"\n📊 Resumen del dataset:")
print(f"   - Total imágenes: {len(dataset)}")
print(f"   - Forma: {dataset.shape}")
print(f"   - Rango de valores: [{dataset.min():.2f}, {dataset.max():.2f}]")

# =====================================================
# CREACIÓN DEL DATASET DE TENSORFLOW
# =====================================================
def create_dataset(images, batch_size):
    """
    Crea un dataset de TensorFlow optimizado
    Args:
        images: Array de imágenes
        batch_size: Tamaño del batch
    Returns:
        tf.data.Dataset
    """
    dataset = tf.data.Dataset.from_tensor_slices(images)
    dataset = dataset.shuffle(1000)
    dataset = dataset.batch(batch_size, drop_remainder=True)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    return dataset

train_dataset = create_dataset(dataset, BATCH_SIZE)

# Verificar que el dataset funcione
print("\n🔍 Verificando dataset...")
for batch in train_dataset.take(1):
    print(f"   - Batch shape: {batch.shape}")
    print(f"   - Batch range: [{batch.numpy().min():.2f}, {batch.numpy().max():.2f}]")

print(f"\n📊 Total batches por epoch: {len(train_dataset)}")
print("=" * 60)
print("✅ CARGA DE IMÁGENES COMPLETADA")
print("=" * 60)
```

```python
# =====================================================
# DEFINICIÓN DEL GENERADOR
# =====================================================
def build_generator():
    """
    Construye el modelo del generador
    Returns:
        Modelo Keras
    """
    model = tf.keras.Sequential([
        # Capa densa inicial
        layers.Dense(4*4*256, input_dim=LATENT_DIM),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Reshape((4, 4, 256)),

        # Capas de deconvolución
        layers.Conv2DTranspose(128, 4, strides=2, padding='same'),
        layers.BatchNormalization(),
        layers.ReLU(),

        layers.Conv2DTranspose(64, 4, strides=2, padding='same'),
        layers.BatchNormalization(),
        layers.ReLU(),

        layers.Conv2DTranspose(32, 4, strides=2, padding='same'),
        layers.BatchNormalization(),
        layers.ReLU(),

        # Capa final
        layers.Conv2DTranspose(CHANNELS, 3, strides=2, padding='same', activation='tanh')
    ])
    return model

# =====================================================
# DEFINICIÓN DEL DISCRIMINADOR
# =====================================================
def build_discriminator():
    """
    Construye el modelo del discriminador
    Returns:
        Modelo Keras
    """
    model = tf.keras.Sequential([
        # Capas convolucionales
        layers.Conv2D(32, 3, strides=2, padding='same', input_shape=(IMG_SIZE, IMG_SIZE, CHANNELS)),
        layers.LeakyReLU(0.2),
        layers.Dropout(0.3),

        layers.Conv2D(64, 3, strides=2, padding='same'),
        layers.LeakyReLU(0.2),
        layers.Dropout(0.3),

        layers.Conv2D(128, 3, strides=2, padding='same'),
        layers.LeakyReLU(0.2),
        layers.Dropout(0.3),

        layers.Conv2D(256, 3, strides=2, padding='same'),
        layers.LeakyReLU(0.2),
        layers.Dropout(0.3),

        # Capa final
        layers.Flatten(),
        layers.Dense(1)  # Sin sigmoid, usamos from_logits=True
    ])
    return model

# =====================================================
# CREACIÓN DE LOS MODELOS
# =====================================================
print("\n" + "=" * 60)
print("🔧 CONSTRUYENDO MODELOS")
print("=" * 60)

generator = build_generator()
discriminator = build_discriminator()

print("\n📊 Generador:")
generator.summary()

print("\n📊 Discriminador:")
discriminator.summary()

# =====================================================
# CONFIGURACIÓN DEL ENTRENAMIENTO
# =====================================================
# Optimizadores
gen_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)
disc_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)

# Función de pérdida
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    return real_loss + fake_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

# =====================================================
# FUNCIÓN DE ENTRENAMIENTO
# =====================================================
@tf.function
def train_step(images):
    """
    Realiza un paso de entrenamiento
    Args:
        images: Batch de imágenes reales
    Returns:
        gen_loss, disc_loss
    """
    noise = tf.random.normal([tf.shape(images)[0], LATENT_DIM])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)

        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    gen_gradients = gen_tape.gradient(gen_loss, generator.trainable_variables)
    disc_gradients = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    gen_optimizer.apply_gradients(zip(gen_gradients, generator.trainable_variables))
    disc_optimizer.apply_gradients(zip(disc_gradients, discriminator.trainable_variables))

    return gen_loss, disc_loss

# =====================================================
# FUNCIONES DE VISUALIZACIÓN (CORREGIDAS)
# =====================================================
def save_generated_images(model, epoch, seed):
    """
    Guarda imágenes generadas
    Args:
        model: Modelo generador
        epoch: Número de epoch (puede ser int o string)
        seed: Semilla de ruido
    """
    os.makedirs('/content/generated_images', exist_ok=True)
    predictions = model(seed, training=False)

    fig = plt.figure(figsize=(8, 8))
    for i in range(min(16, predictions.shape[0])):
        plt.subplot(4, 4, i+1)
        plt.imshow(predictions[i] * 0.5 + 0.5)
        plt.axis('off')

    # CORRECCIÓN: Verificar si epoch es string o número
    if isinstance(epoch, str):
        plt.savefig(f'/content/generated_images/epoch_{epoch}.png')
    else:
        plt.savefig(f'/content/generated_images/epoch_{epoch:03d}.png')

    plt.close()

def plot_combined_losses(gen_losses, disc_losses):
    """
    Grafica la evolución de las pérdidas del generador y discriminador juntas
    Args:
        gen_losses: Lista de pérdidas del generador
        disc_losses: Lista de pérdidas del discriminador
    """
    # Crear figura con tamaño adecuado
    plt.figure(figsize=(12, 6))

    # Graficar ambas pérdidas en la misma figura
    epochs = range(1, len(gen_losses) + 1)

    plt.plot(epochs, gen_losses, label='Generator Loss', color='blue', linewidth=2, marker='o', markersize=4)
    plt.plot(epochs, disc_losses, label='Discriminator Loss', color='red', linewidth=2, marker='s', markersize=4)

    # Configurar el gráfico
    plt.title('Evolución de las Pérdidas del Generador y Discriminador', fontsize=14, fontweight='bold')
    plt.xlabel('Épocas', fontsize=12)
    plt.ylabel('Pérdida', fontsize=12)
    plt.legend(loc='best', fontsize=11)
    plt.grid(True, alpha=0.3)

    # Añadir líneas de referencia
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.1)

    # Ajustar límites del eje y para mejor visualización
    all_losses = gen_losses + disc_losses
    y_max = max(all_losses) * 1.1 if all_losses else 1
    plt.ylim(0, y_max)

    # Añadir anotaciones de valores finales
    if gen_losses and disc_losses:
        plt.annotate(f'Gen: {gen_losses[-1]:.3f}',
                    xy=(len(gen_losses), gen_losses[-1]),
                    xytext=(len(gen_losses)*0.7, gen_losses[-1]*0.5),
                    arrowprops=dict(arrowstyle='->', color='blue'),
                    color='blue', fontsize=10)

        plt.annotate(f'Disc: {disc_losses[-1]:.3f}',
                    xy=(len(disc_losses), disc_losses[-1]),
                    xytext=(len(disc_losses)*0.7, disc_losses[-1]*1.3),
                    arrowprops=dict(arrowstyle='->', color='red'),
                    color='red', fontsize=10)

    plt.tight_layout()
    plt.savefig('/content/combined_losses.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Imprimir estadísticas resumidas
    print("\n📊 Estadísticas de pérdidas:")
    print(f"   - Generator Loss final: {gen_losses[-1]:.4f}" if gen_losses else "   - Sin datos de generador")
    print(f"   - Discriminator Loss final: {disc_losses[-1]:.4f}" if disc_losses else "   - Sin datos de discriminador")
    print(f"   - Generator Loss mínimo: {min(gen_losses):.4f}" if gen_losses else "")
    print(f"   - Discriminator Loss mínimo: {min(disc_losses):.4f}" if disc_losses else "")

def plot_losses_separate(gen_losses, disc_losses):
    """
    Opcional: Grafica las pérdidas en gráficos separados (para comparación)
    Args:
        gen_losses: Lista de pérdidas del generador
        disc_losses: Lista de pérdidas del discriminador
    """
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(gen_losses, label='Generator', color='blue')
    plt.title('Generator Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(disc_losses, label='Discriminator', color='red')
    plt.title('Discriminator Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('/content/training_losses_separate.png')
    plt.show()

# =====================================================
# ENTRENAMIENTO
# =====================================================
print("\n" + "=" * 60)
print("🚀 INICIANDO ENTRENAMIENTO")
print("=" * 60)
print(f"📊 Epochs: {EPOCHS}")
print(f"📊 Batch size: {BATCH_SIZE}")
print(f"📊 Total batches: {len(train_dataset)}")
print("⏱️  Presiona Ctrl+C para detener\n")

# Semilla fija para generar imágenes consistentes
seed = tf.random.normal([16, LATENT_DIM])

# Listas para almacenar pérdidas
gen_losses = []
disc_losses = []
start_time = time.time()

# Bucle de entrenamiento
for epoch in range(EPOCHS):
    epoch_start = time.time()
    epoch_gen_losses = []
    epoch_disc_losses = []

    # Barra de progreso
    pbar = tqdm(train_dataset, desc=f'Epoch {epoch+1}/{EPOCHS}')

    for batch in pbar:
        try:
            # Realizar paso de entrenamiento
            gen_loss, disc_loss = train_step(batch)

            # Convertir a numpy
            gen_loss_val = gen_loss.numpy()
            disc_loss_val = disc_loss.numpy()

            # Almacenar pérdidas
            epoch_gen_losses.append(gen_loss_val)
            epoch_disc_losses.append(disc_loss_val)

            # Actualizar barra de progreso
            pbar.set_postfix({
                'Gen': f'{gen_loss_val:.3f}',
                'Disc': f'{disc_loss_val:.3f}'
            })

        except Exception as e:
            print(f"❌ Error en batch: {e}")
            continue

    # Calcular promedios del epoch
    avg_gen = np.mean(epoch_gen_losses) if epoch_gen_losses else 0
    avg_disc = np.mean(epoch_disc_losses) if epoch_disc_losses else 0

    gen_losses.append(avg_gen)
    disc_losses.append(avg_disc)

    epoch_time = time.time() - epoch_start

    print(f"📊 Epoch {epoch+1}: Gen={avg_gen:.3f}, Disc={avg_disc:.3f}, Tiempo={epoch_time:.1f}s")

    # Guardar imágenes cada 5 epochs
    if (epoch + 1) % 5 == 0:
        save_generated_images(generator, epoch+1, seed)
        print(f"   📸 Imágenes guardadas para epoch {epoch+1}")

# =====================================================
# FINALIZACIÓN
# =====================================================
total_time = time.time() - start_time
print("\n" + "=" * 60)
print("✅ ENTRENAMIENTO COMPLETADO")
print("=" * 60)
print(f"⏱️  Tiempo total: {total_time/60:.1f} minutos")

# Graficar pérdidas COMBINADAS en una sola gráfica
print("\n📊 Generando gráfica combinada de pérdidas...")
plot_combined_losses(gen_losses, disc_losses)

# Guardar imágenes finales - CORREGIDO: usamos un número en lugar de "final"
print("\n🎨 Generando imágenes finales...")
save_generated_images(generator, EPOCHS, seed)  # Usamos EPOCHS como número

# =====================================================
# GUARDAR MODELO
# =====================================================
try:
    generator.save('/content/generator_model.h5')
    print("✅ Modelo generador guardado en /content/generator_model.h5")
except Exception as e:
    print(f"⚠️ Error guardando modelo: {e}")

# =====================================================
# GENERAR MÚLTIPLES ROSTROS
# =====================================================
def generate_faces(model, num_faces=20):
    """
    Genera múltiples rostros
    Args:
        model: Modelo generador
        num_faces: Número de rostros a generar
    """
    os.makedirs('/content/generated_faces', exist_ok=True)
    batch_size = 8

    for i in range(0, num_faces, batch_size):
        actual_size = min(batch_size, num_faces - i)
        noise = tf.random.normal([actual_size, LATENT_DIM])
        faces = model(noise, training=False)

        for j in range(actual_size):
            img = (faces[j].numpy() * 0.5 + 0.5) * 255
            img = np.clip(img, 0, 255).astype(np.uint8)
            Image.fromarray(img).save(f'/content/generated_faces/face_{i+j+1:03d}.png')

    print(f"✅ {num_faces} rostros generados en /content/generated_faces/")

print("\n🎨 Generando múltiples rostros...")
generate_faces(generator, 20)

print("\n" + "=" * 60)
print("🎉 ¡PROCESO COMPLETADO EXITOSAMENTE!")
print("=" * 60)
print("\n📁 Archivos generados:")
print("   - /content/generated_images/ (imágenes de progreso)")
print("   - /content/generated_faces/ (rostros finales)")
print("   - /content/generator_model.h5 (modelo guardado)")
print("   - /content/combined_losses.png (gráfica combinada de pérdidas)")
print("\n" + "=" * 60)
```

```python
# =====================================================
# 3. FUNCIONES DE PROCESAMIENTO PARA GAN
# =====================================================

def preprocesar_para_gan(imagen_path):
    """
    Preprocesa la imagen para el modelo GAN
    Args:
        imagen_path: Ruta de la imagen
    Returns:
        Imagen preprocesada (64, 64, 3)

    """
    IMG_SIZE=64
    # Cargar imagen
    img = Image.open(imagen_path)

    # Convertir a RGB
    if img.mode != 'RGB':
        img = img.convert('RGB')

    # Redimensionar a 64x64
    img_resized = img.resize((IMG_SIZE, IMG_SIZE), Image.Resampling.LANCZOS)

    # Normalizar a [-1, 1]
    img_array = np.array(img_resized, dtype=np.float32) / 127.5 - 1.0

    return img_array

def generar_rostro_gan(modelo, imagen_preprocesada):
    """
    Genera un rostro usando el modelo GAN
    Args:
        modelo: Modelo GAN cargado
        imagen_preprocesada: Imagen preprocesada (64, 64, 3)
    Returns:
        Imagen generada (64, 64, 3)
    """
    # Añadir dimensión de batch
    imagen_batch = np.expand_dims(imagen_preprocesada, axis=0)

    # Generar imagen
    try:
        imagen_generada = modelo.predict(imagen_batch, verbose=0)
    except:
        # Si falla, generar desde ruido
        print("⚠️ Generando desde ruido aleatorio...")
        noise = tf.random.normal([1, LATENT_DIM])
        imagen_generada = modelo.predict(noise, verbose=0)

    return imagen_generada[0]
```
